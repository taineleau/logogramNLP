```python
import json
```

```python
collection_id = '1-wenshu' # '3-book' # '2-divination' #
data = json.load(open(f"/Users/danluchen/beeleaf/jiandu/baoshan/{collection_id}_wordmap.json"))
```

```python
# all_slips = {}
for x in data['wordmap']:
    all_slips[f"{collection_id}_{x}"] = data['wordmap'][x]
```

```python
print(len(all_slips))
```

```python
import random
cnt = 0
train_valid_record = {}
for x in all_slips:
    if random.random() < 0.9:
        train_valid_record[x] = True # is training set
        cnt += 1
    else:
        train_valid_record[x] = False
```

```python
train_dict = {'sentence': [], 'label': [], 'raw_sent': []}
valid_dict = {'sentence': [], 'label': [], 'raw_sent': []}

for x in train_valid_record:
    if train_valid_record[x]:
        train_dict['sentence'].append("".join(all_slips[x]))
        train_dict['raw_sent'].append(all_slips[x])
        train_dict['label'].append(x.split("_")[0])
    else:
        valid_dict['sentence'].append("".join(all_slips[x]))
        valid_dict['raw_sent'].append(all_slips[x])
        valid_dict['label'].append(x.split("_")[0])
```

```python
train_dict = []
valid_dict = [] #{'sentence': [], 'label': []}

for x in train_valid_record:
    if train_valid_record[x]:
        tmp = {'sentence': "".join(all_slips[x]), 'label': x.split("_")[0]}
        train_dict.append(tmp)
    else:
        tmp = {'sentence': "".join(all_slips[x]), 'label': x.split("_")[0]}
        valid_dict.append(tmp)
        
json.dump(train_dict, open("train.json", "w"))
json.dump(valid_dict, open("valid.json", "w"))
```
