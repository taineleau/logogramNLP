```python
import glob
```

```python
# folder_id = "01-laozijia"
folder_id = ["01-laozijia",
             "02-laoziyi",
             '03-laozibing',
             "04-taiyishengshui",
             '05-liuyi',
             '06-wuxing',
             '07'][6]

all_imgs = glob.glob(f"./{folder_id}/*.png")
```

```python
import os
img_map = {}
for i_path in all_imgs:
    file_name = i_path.split("/")[-1]
    if '-' in file_name:
        img_id, split_id = file_name.strip(".png").split('-')
        if img_id not in img_map:
            img_map[img_id] = {}
        img_map[img_id][int(split_id)] = i_path
    else:
        img_id = file_name.strip(".png")
        img_map[img_id] = {}
        img_map[img_id]['full'] = i_path
```

```python
from PIL import Image
base_path = '/Users/danluchen/beeleaf/jiandu/training_data/translation_478/pixel_imgs_24height_handcopy'

global_height = 24
for img_id in sorted(img_map.keys()):
    full_length = 0
    im_lists = []
    # print(img_id)
    # break
    if 'full' not in img_map[img_id]:
        keys = sorted(img_map[img_id].keys())

        for key in keys:
            i_path = img_map[img_id][key]
            im = Image.open(i_path)
            im = im.resize((int(im.size[0] / im.size[1] * global_height), global_height))
            full_length += im.size[0]
            im_lists.append(im)
    else:
        im = Image.open(img_map[img_id]['full'])
        im = im.resize((int(im.size[0] / im.size[1] * global_height), global_height))
        im_lists = [im]
        full_length = im.size[0]
    
    ## concat     
    line_image = Image.new("L", size=(full_length, global_height), color=255)
    num_patches = (full_length - 1) // global_height + 1
    
    offset = 0
    for im in im_lists:
        line_image.paste(im, (offset, 0))
        offset += im.size[0]
    print(line_image.size)

    # black_eos = Image.new("L", size=(16, 16), color=0)
    # line_image.paste(black_eos, (num_patches*16, 0))
    display(line_image)
    line_image.save(os.path.join(base_path, f"{folder_id}_{img_id}.png"))
    print(os.path.join(base_path, f"{folder_id}_{img_id}.png"))
    
        # break
```
